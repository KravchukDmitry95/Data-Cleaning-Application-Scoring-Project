```python
import pandas as pd 
df = pd.read_csv('C:/Anaconda/Lib/applications(2.0).csv')
df.drop_duplicates(subset = 'applicant_id')
df['External Rating'] = df['External Rating'].fillna(0)
df['Education level'] = df['Education level'].fillna('Середня')
df2 = pd.read_csv('C:/Anaconda/Lib/industries(2.0).csv')
df = df.merge(df2, on= 'Industry', how='left')
from datetime import datetime
df['Applied at'] = pd.to_datetime(df['Applied at'], format ='mixed')
def calculated_rate(row):
    rating = 0 
    if pd.isna(row['Amount']) or row['External Rating'] == 0:
        return 0;
    if 35 <= row['Age'] <= 55:
        rating += 20
    if row['Applied at'].weekday() not in [5,6]:
        rating += 20
    if row['Marital status'] == 'Married':
        rating += 20
    if row['Location'] == 'Київ чи область':
        rating += 10
    rating += row['Score']
    if row['External Rating'] >= 7:
        rating += 20
    if row['External Rating'] <= 2:
        rating -= 20 
    rating = min(rating, 100)
    return rating
df['Rating'] = df.apply(calculated_rate, axis = 1)
df = df[df['Rating'] > 0]
df['Week'] = df['Applied at'].dt.isocalendar().week 
weekly_avg = df.groupby('Week')['Rating'].mean()
print (df)


        


```
